```python
import pandas as pd
import numpy as np
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(128, input_shape=(1,10), return_sequences = True))
model.add(LSTM(128))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
model.load_weights('model_oversampled250.h5')
```

```python
from tkinter import *
root=Tk()
#root.geometry("1200x500")
HR=StringVar()
O2Sat=StringVar()
Temp=StringVar()
SBP=StringVar()
MAP=StringVar()
Resp=StringVar()
Age=StringVar()
Gender = IntVar()
HospAdmTime=StringVar()
ICULOS=StringVar()
pred1=StringVar()

def Reset():
    HR.set("")
    O2Sat.set("")
    Temp.set("")
    SBP.set("")
    MAP.set("")
    Resp.set("")
    Gender.set(None)
    Age.set("")
    HospAdmTime.set("")
    ICULOS.set("")
    pred1.set("")

    
from tkinter import messagebox
def Exit():
    qExit=messagebox.askyesno("System","Do you want to exit the system")
    
    if qExit:
        root.destroy()
        exit()

def classify(x):
    if x <= 0.5:
        return 0
    else:
        return 1 

def sepsis():
    hr=HR.get()
    o2sat=O2Sat.get()
    temp=Temp.get()
    sbp=SBP.get()
    map1=MAP.get()
    resp=Resp.get()
    if(Gender.get()==1):
        gender=1.0
    else:
        gender=0.0
    age=Age.get()
    hospadmtime=HospAdmTime.get()
    iculos=ICULOS.get()
    X = [float(hr),float(o2sat),float(temp),float(sbp),float(map1),float(resp),float(gender),float(age),float(hospadmtime),float(iculos)]
    X_mean = [83.72276626338754, 96.83018891312972,36.61446499405063,123.04661877727092,82.62690784906772,18.90137532229274,61.64342324474379,0.5594506148353828,-51.84765965886919,39.01011503371678]
    X_std = [16.5302153248836,3.913998151342051,3.1449394226042457,24.17546288602834,16.42480213606982,4.791598816912431,16.48294561381705,0.4964591975794707,139.7649578409348,22.924640812712394]
    X = np.array(X)
    X_mean = np.array(X_mean)
    X_std = np.array(X_std)
    X=((X-X_mean)/X_std)
    X = X.reshape(1,1,10)
    y = model.predict(X)
    final_pred = np.array(list(map(classify, y)))
    if final_pred[0]==0:
        pred1.set("You don't have Sepsis")
    else:
        pred1.set("You have Sepsis")
    
#Tk class is used to create a root window
root.configure(background='cyan')
root.title('Sepsis Prediction')
root.resizable(0,0)

#Headings for the GUI written at the top of GUI
w2 = Label(root, justify=LEFT, text="Sepsis Prediction Using Vital Signs and Demographics", fg="Red", bg="cyan")
w2.config(font=("Times",30,"bold italic"))
w2.grid(row=1, column=0, columnspan=2, padx=100)
w2 = Label(root, justify=LEFT, text="Contributors: Kunal, Keshav and Vedant", fg="Blue", bg="cyan")
w2.config(font=("Times",30,"bold italic"))
w2.grid(row=2, column=0, columnspan=2, padx=100)

#Creating Labels for the symtoms
S1Lb = Label(root, text="HR", fg="Black", bg="cyan")
S1Lb.config(font=("Times",15,"bold italic"))
S1Lb.grid(row=7, column=0, pady=10, sticky=W)

name_entry_HR = Entry(root,textvariable = HR, font=('calibre',10,'normal'))
name_entry_HR.grid(row=7,column=0)

S2Lb = Label(root, text="O2Sat", fg="Black", bg="cyan")
S2Lb.config(font=("Times",15,"bold italic"))
S2Lb.grid(row=8, column=0, pady=10, sticky=W)

name_entry_O2Sat = Entry(root,textvariable = O2Sat, font=('calibre',10,'normal'))
name_entry_O2Sat.grid(row=8,column=0)

S3Lb = Label(root, text="Temp", fg="Black",bg="cyan")
S3Lb.config(font=("Times",15,"bold italic"))
S3Lb.grid(row=9, column=0, pady=10, sticky=W)

name_entry_Temp = Entry(root,textvariable = Temp, font=('calibre',10,'normal'))
name_entry_Temp.grid(row=9,column=0)

S4Lb = Label(root, text="SBP", fg="Black", bg="cyan")
S4Lb.config(font=("Times",15,"bold italic"))
S4Lb.grid(row=10, column=0, pady=10, sticky=W)

name_entry_SBP = Entry(root,textvariable = SBP, font=('calibre',10,'normal'))
name_entry_SBP.grid(row=10,column=0)

S5Lb = Label(root, text="MAP", fg="Black", bg="cyan")
S5Lb.config(font=("Times",15,"bold italic"))
S5Lb.grid(row=11, column=0, pady=10, sticky=W)

name_entry_MAP = Entry(root,textvariable = MAP, font=('calibre',10,'normal'))
name_entry_MAP.grid(row=11,column=0)

S5Lb = Label(root, text="Resp", fg="Black", bg="cyan")
S5Lb.config(font=("Times",15,"bold italic"))
S5Lb.grid(row=12, column=0, pady=10, sticky=W)

name_entry_resp = Entry(root,textvariable = Resp, font=('calibre',10,'normal'))
name_entry_resp.grid(row=12,column=0)

S7Lb = Label(root, text="Age", fg="Black", bg="cyan")
S7Lb.config(font=("Times",15,"bold italic"))
S7Lb.grid(row=13, column=0, pady=10, sticky=W)

name_entry_age = Entry(root,textvariable = Age, font=('calibre',10,'normal'))
name_entry_age.grid(row=13,column=0)

S6Lb = Label(root, text="Gender", fg="Black", bg="cyan")
S6Lb.config(font=("Times",15,"bold italic"))
S6Lb.grid(row=14, column=0, pady=10, sticky=W)

male = Radiobutton(root, text="Male", variable=Gender,value=1)
#male.pack( anchor = W )
male.grid(row=14,column=0)

female = Radiobutton(root, text="Female", variable=Gender,value=2)
#female.pack( anchor = W )
female.grid(row=15,column=0)

S8Lb = Label(root, text="HospAdmTime", fg="Black", bg="cyan")
S8Lb.config(font=("Times",15,"bold italic"))
S8Lb.grid(row=16, column=0, pady=10, sticky=W)

name_entry_hosp = Entry(root,textvariable = HospAdmTime, font=('calibre',10,'normal'))
name_entry_hosp.grid(row=16,column=0)

S9Lb = Label(root, text="ICULOS", fg="Black", bg="cyan")
S9Lb.config(font=("Times",15,"bold italic"))
S9Lb.grid(row=17, column=0, pady=10, sticky=W)

name_entry_ICULOS = Entry(root,textvariable = ICULOS, font=('calibre',10,'normal'))
name_entry_ICULOS.grid(row=17,column=0)


dst = Button(root, text="Predict Sepsis",command=sepsis, bg="darkviolet",fg="white")
dst.config(font=("Times",15,"bold italic"))
dst.grid(row=8, column=1,padx=10)

#Showing the output of different aldorithms
t1=Label(root,font=("Times",15,"bold italic"),text="Predicted Disaese",height=1,bg="white"
         ,width=20,fg="darkviolet",textvariable=pred1,relief="sunken").grid(row=9, column=1, padx=10)


rs = Button(root,text="Reset Inputs", command=Reset,bg="red",fg="white",width=15)
rs.config(font=("Times",15,"bold italic"))
rs.grid(row=11,column=1,padx=10)

ex = Button(root,text="Exit", command=Exit,bg="red",fg="white",width=15)
ex.config(font=("Times",15,"bold italic"))
ex.grid(row=12,column=1,padx=10)

#calling this function because the application is ready to run
root.mainloop()
```
